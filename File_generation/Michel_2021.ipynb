```python

# for every row in the df, find the correlating NAA cells
# for every row in dataframe, find the griddcell index in the dataset
from gridcellfinder import gridcellfinder as gridcellfinder

def LonLat_cleaner(df): # this function differs from the other function :\
  list_of_lons = []   
  list_of_lats = []
  # I need to include a switch for nan's
  # I need to include a switch for east/west
  for x in df['Latitude']:
       # this code is from Nic Anau '''
       if "'" not in x:
             y = float(x[:2])+float(x[3:])/60
       else:
             y = float(x[:2])+float(x[3:-1])/60
       list_of_lats.append(y)
  for x in df['Longitude']:
       # this code is from Nic Anau '''
       if "'" not in x:
             z = float(x[:3])+float(x[4:])/60
       else:
             z = float(x[:3])+float(x[4:-1])/60
       
       if z > 0: # account for some coordinates being positive
            z = z *-1     
       else: 
             z = z         
       list_of_lons.append(z)
  
     
       
  df['Longitude'] = list_of_lons
  df['Latitude'] = list_of_lats
  return(df) 

```

```python
df18 = LonLat_cleaner(df18)
df19 = LonLat_cleaner(df19)
```

```python
# run the function for every row in the dataframe
gcl = pd.DataFrame( columns=['X', 'Y'] )
df['Longitude'] = df['Longitude'].astype(float)
df['Latitude'] = df['Latitude'].astype(float)

df['lons'] = df['Longitude'].round(2) 
df['lats'] = df['Latitude'].round(2)
df['lonmin'] = df['Longitude'].round(decimals=2) - 0.1 #0.1 in degrees is appox 10 km -> i read in  Roberts(?) 2022 paper, they did the same
df['lonmax'] = df['Longitude'].round(decimals=2) + 0.1
df['latmin'] = df['Latitude'].round(decimals=2) - 0.1
df['latmax'] = df['Latitude'].round(decimals=2) + 0.1

for i in [*range(0,len(df),1)]: 
  x,y = gridcellfinder(df, xrds, i)
  X = x.tolist()
  Y = y.tolist()
  gcl.loc[len(gcl.index)] = [X, Y] 
# This x andy are the grid-coordinates for the NAA grid.   
 
if depth == True: 
    ncds = xr.open_dataset('DATA/NAA_Area_Volume.nc') # this is the updated file with corrected depth cut-offs
    z = [0]
    zc  = list(ncds.z_cuts.values)
    zcu = z + zc
    deptht = ncds.deptht.values
    df['NAA_deptht'] = pd.cut(df['Depth [m]'], bins = zcu , labels =  deptht)
else:
  print( ':)' )    
    
 
```
