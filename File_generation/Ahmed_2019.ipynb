```python
# for every row in dataframe, find the griddcell index in the dataset
def gridcellfinder(df, xrds, i): 
  '''returns all the gridcell indices for row i as single number'''
  mask = xrds['nav_lon'].where((xrds['nav_lat'] > df['latmin'][i]) & (xrds['nav_lat'] < df['latmax'][i]) &
                           (xrds['nav_lon'] > df['lonmin'][i]) & (xrds['nav_lon'] < df['lonmax'][i]),0)
  fu = np.array(mask)
  fu2 = np.where(fu != 0)
  X = fu2[0][:]
  Y = fu2[1][:]
  
  return X, Y
```

```python
# try to extrace the datetim eas single column, maybe I need to do this by hand
times = df.datetime.str.split(' ', expand=True)
# get list of all dates 
day_list = list(times[0].unique()) 
print(len(day_list))  
```

```python
# Replace values in the day_list
new_day_list = []
for i in range(len(day_list)):
  if '/' in day_list[i]:  # checks if the datestams contains has a / in it. 
    new_day_list.append(day_list[i].split('/')) # split the string -> separator is /
  else:
    new_day_list.append(day_list[i].split('-')) # split the string -> separator is -

for i in  range(len(day_list)):
  # check the length of year string: 
  if len(new_day_list[i][2]) == 2:
    new_day_list[i][2] = '20' + new_day_list[i][2]
  #check the length of day string:
  if len(new_day_list[i][0]) == 1:
    new_day_list[i][0] = '0' + new_day_list[i][0]
  #check the length of month string:
  if len(new_day_list[i][1]) == 1:
    new_day_list[i][1] = '0' + new_day_list[i][1]
#new_day_list  

## tadaa: new list object with uniform format. 
```

```python
# bind the new list together to YYYY/MM/DD
replacement_list = []
for i in range(len(new_day_list)):
  replacement_list.append(new_day_list[i][2] + '/' + new_day_list[i][1] + '/' + new_day_list[i][0])
```

```python
# replacing the values with the dataframe
time = times[0].replace(to_replace=day_list,value= replacement_list)
timestamp = times[1]
day = pd.to_datetime(time, format='%Y/%m/%d')
```
